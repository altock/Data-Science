```python
# L5, S38

import pandas as pd

r_cols = ['user_id', 'movie_id', 'rating']
ratings = pd.read_csv('DataScience-Python3/ml-100k/u.data', sep='\t',
                     names=r_cols, usecols=range(3), encoding="ISO-8859-1")

m_cols = ['movie_id', 'title']
movies = pd.read_csv('DataScience-Python3/ml-100k/u.item', sep='|',
                     names=m_cols, usecols=range(2), encoding="ISO-8859-1")

# TODO: Use this in my recommender? Will combine anime + ratings
ratings = pd.merge(movies, ratings)
ratings.head()
```

```python
# TODO: Also seems useful for calculating similarities
movieRatings = ratings.pivot_table(index=['user_id'], columns='title', values='rating')
movieRatings.head()
```

```python
starWarsRatings = movieRatings['Star Wars (1977)']
starWarsRatings.head()
```

```python
similarMovies = movieRatings.corrwith(starWarsRatings)
similarMovies = similarMovies.dropna()
df = pd.DataFrame(similarMovies)
df.head(10)
```

```python
similarMovies.order(ascending=False)
```

```python
import numpy as np
movieStats = ratings.groupby('title').agg({'rating': [np.size, np.mean]})
movieStats.head()
```

```python
popularMovies = movieStats['rating']['size'] >= 40
movieStats[popularMovies].sort_values([('rating', 'mean')], ascending=False)[:15]
```

```python
df = movieStats[popularMovies].join(pd.DataFrame(similarMovies, columns=['similarity']))
```

```python
df.head()
```

```python
df.sort_values(['similarity'], ascending=False)[:15]
```

```python
movieRatings.head()
```

```python
# L5, S40
userRatings = movieRatings
```

```python
corrMatrix = userRatings.corr()
corrMatrix.head()
```

```python
corrMatrix = corrMatrix.corr(method='pearson', min_periods=100)
corrMatrix.head()
```

```python
myRatings = userRatings.loc[0].dropna()
myRatings
```

```python
simCandidates = pd.Series()
for i in range(0, len(myRatings.index)):
    print("Adding sims for " + myRatings.index[i] + "...")
    # Retrieve similar movies to this one that I rated
    sims = corrMatrix[myRatings.index[i]].dropna()
    # Now scale its similarity by how well I rated this movie
    sims = sims.map(lambda x: x * myRatings[i])
    # Add the score to the list of similarity candidates
    simCandidates = simCandidates.append(sims)
#Glance at our results so far:
print("sorting...")
simCandidates.sort_values(inplace=True, ascending=False)
simCandidates.head(10)
```

```python
simCandidates = simCandidates.groupby(simCandidates.index).sum()
```

```python
simCandidates.sort_values(inplace=True, ascending=False)
simCandidates.head(10)
```

```python
simCandidates = simCandidates.drop(myRatings.index)
simCandidates.head(10)
```
