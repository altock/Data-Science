```python
np.percentile(vals, 50)
```

```python
np.percentile(vals, 99)
```

```python
np.percentile(vals, 10)
```

```python
# mu_n = \int_{-\inf}^\inf (x - c)^n f(x) dx for some c
# First moment is the mean
# Second moment is the variance
# Third moment is "skew", how lopsided is the distribution? longer tail on left -> skewed left/negative skew
# Fourth moment is "kurtosis", how thick is the tail, and how sharp the peak compared to normal distribution.
# Higher peaks have higher kurtosis
```

```python
# Should be around 0
np.mean(vals)
```

```python
# Should be std**2, so 0.5**2 = 0.25
np.var(vals)
```

```python
import scipy.stats as sp
# Should be around 0 since data isn't lopsided
sp.skew(vals)
```

```python
# Should be around 0 since this is the normal distribution
sp.kurtosis(vals)
```

```python
# S2, L15

def de_mean(x):
    xmean = np.mean(x)
    return [xi - xmean for xi in x]

def covariance(x, y):
    n = len(x)
    # Sample variance
    return np.dot(de_mean(x), de_mean(y)) / (n-1)

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = np.random.normal(50.0, 10.0, 1000)

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()
covariance(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = np.random.normal(50.0, 10.0, 1000) / pageSpeeds

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()
covariance(pageSpeeds, purchaseAmount)
```

```python
# Pearson correlation
def correlation(x,y):
    std_x = x.std()
    std_y = y.std()
    return covariance(x,y) / (std_x * std_y)

correlation(pageSpeeds, purchaseAmount)
```

```python
np.corrcoef(pageSpeeds, purchaseAmount)
```

```python
purchaseAmount = 100 - pageSpeeds * 3

plt.scatter(pageSpeeds, purchaseAmount)
plt.show()
print(covariance(pageSpeeds, purchaseAmount), correlation(pageSpeeds, purchaseAmount))
```
